# Author Desambiguation

```python
#
# Data loading
#
import pandas as pd

df = pd.read_json(
    "https://raw.githubusercontent.com/jdvelasq/techminer/master/data/tutorial/"
    + "author-desambiguation.json",
    orient="records",
    lines=True,
)
```

```python
#
# A column for identify each record is added to the dataframe.
#
from techminer import DataFrame

df = DataFrame(df).disambiguate_authors()
```
